```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Load datasets from venv37 folder ---
clinical = pd.read_csv("venv37/clinical_breast_cancer.csv")
recurrence = pd.read_csv("venv37/breast_cancer_recurrence.csv")
diagnostic = pd.read_csv("venv37/breast_cancer_diagnostic.csv")

# --- Clean Clinical Dataset ---
if 'Status' in clinical.columns:
    clinical['Status'] = clinical['Status'].map({'Alive': 0, 'Dead': 1})
    clinical = clinical.rename(columns={'Status': 'target'})
clinical['source'] = 'clinical'

# --- Clean Recurrence Dataset ---
if 'Class' in recurrence.columns:
    recurrence['Class'] = recurrence['Class'].map({'no-recurrence-events': 0, 'recurrence-events': 1})
    recurrence = recurrence.rename(columns={'Class': 'target'})
elif 'diagnosis' in recurrence.columns:
    recurrence['diagnosis'] = recurrence['diagnosis'].map({'M': 1, 'B': 0})
    recurrence = recurrence.rename(columns={'diagnosis': 'target'})
recurrence['source'] = 'recurrence'

# --- Clean Diagnostic Dataset ---
if 'diagnosis' in diagnostic.columns:
    diagnostic['diagnosis'] = diagnostic['diagnosis'].map({'M': 1, 'B': 0})
    diagnostic = diagnostic.rename(columns={'diagnosis': 'target'})
diagnostic['source'] = 'diagnostic'

# --- Concatenate datasets ---
combined = pd.concat([clinical, recurrence, diagnostic], axis=0, ignore_index=True)

# --- Save combined dataset back into venv37 ---
combined.to_csv("venv37/combined.csv", index=False)

print("Combined dataset saved as venv37/combined.csv")
print("Shape:", combined.shape)

# --- Preview first 10 rows ---
display(combined.head(10))

# --- Summary: count rows from each source ---
print("\nRow counts by dataset source:")
print(combined['source'].value_counts())

# --- Statistical summary of numeric columns ---
print("\nStatistical summary of numeric features:")
display(combined.describe())

# --- Correlation heatmap (Matplotlib only) ---
corr = combined.corr(numeric_only=True)

plt.figure(figsize=(12,8))
plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap of Combined Dataset", fontsize=14)
plt.show()

# --- Sorted correlations with target (Matplotlib only) ---
target_corr = corr['target'].drop('target').sort_values(ascending=False)

plt.figure(figsize=(10,6))
plt.barh(target_corr.index, target_corr.values, color="skyblue")
plt.title("Feature Correlations with Target", fontsize=14)
plt.xlabel("Correlation")
plt.ylabel("Feature")
plt.gca().invert_yaxis()  # highest correlation at top
plt.show()
```
